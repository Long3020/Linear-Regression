```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Dữ liệu
x = np.array([[30, 40, 60, 80, 90]]).T #Diện tích
y = np.array([[1500, 1800, 2500, 3200, 3500]]).T #Giá nhà
```

```python
plt.plot(x, y, 'ro')
plt.axis([0, 150, 500, 5000])
plt.xlabel('Diện tích(m2)')
plt.ylabel('Giá nhà(triệu VND)')
plt.show()
```

```python
one = np.ones((x.shape[0], 1))
X = np.concatenate((one, x), axis = 1)

print(one, X)
```

```python
a = np.dot(X.T, X)
b = np.dot(X.T, y)
w = np.dot(np.linalg.pinv(a), b)

print(a, b)
print('w = {}'.format(w))  # Dùng format()
```

```python
w0 = w[0][0]
w1 = w[1][0]
x0 = np.linspace(30, 90, 2, endpoint = True)
y0 = w0 + w1*x0

print(w0, w1, x0, y0)
```

```python
plt.plot(x, y, 'ro')
plt.plot(x0, y0)
plt.axis([0, 150, 500, 5000])
plt.xlabel('Diện tích(m2)')
plt.ylabel('Giá nhà(triệu VND)')
plt.show()
```

```python
y1 = w1*50 + w0
y2 = w1*70 + w0

print(y1, y2)
```

```python
from sklearn import datasets, linear_model

# fit the model by Linear Regression
regr = linear_model.LinearRegression(fit_intercept=False) # fit_intercept = False for calculating the bias
regr.fit(X, y)

# Compare two results
print('Nghiệm tìm được bằng scikit-learn  : ', regr.coef_) 
print ('Nghiệm tìm được từ phương trình (5): ', w.T)
```
